```python
train_data = pd.read_csv("/kaggle/input/playground-series-s3e2/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/playground-series-s3e2/test.csv")
test_data.head()
```

```python
nev = train_data.loc[train_data.smoking_status == "never smoked"]["stroke"]
nv = sum(nev)/len(nev)
print("% of never smoked who had stroke : ",nv)
```

```python
smk = train_data.loc[train_data.smoking_status == "smokes"]["stroke"]
sk = sum(smk)/len(smk)
print("% of smokes who had stroke : ",sk)
```

```python
fms = train_data.loc[train_data.smoking_status == "formerly smoked"]["stroke"]
fm = sum(fms)/len(fms)
print("% of never smoked who had stroke : ",fm)
```

```python
train_data['age'] = train_data['age'].apply(lambda x: 1 if x > 50 else 0)
test_data['age'] = test_data['age'].apply(lambda x: 1 if x > 50 else 0)
```

```python
from sklearn.ensemble import RandomForestRegressor

y = train_data["stroke"]

features = ["smoking_status", "hypertension", "heart_disease", "age", "gender"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestRegressor(n_estimators=100, max_depth=6, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'id': test_data.id, 'stroke': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
